```python
#for activating the webcam

capture=cv.VideoCapture("./videos/dog.mp4")

while True:

    isTrue,frame=capture.read()
    if isTrue:
         cv.imshow("dog_video",frame)
         if cv.waitKey(20) & 0xFF==ord('d'):
                break
   
capture.release()
cv.destroyAllWindows()
         
```

```python
#resizing and rescaling image with opencv


def resize(frame,scale=0.25):
     
    height=int(frame.shape[0]*scale)
    width=int(frame.shape[1]*scale)

    return cv.resize(frame,(width,height))

resized_frame=resize(img1)

cv.imshow("resized_image",resized_frame)

cv.waitKey(0)
```

```python
# converting an image to grayscale

image=cv.imread("./Photos/park.jpg")
cv.imshow("park",image)

# convert image to grayscale
image_gray=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv.imshow("park_gray",image_gray)

# flouter une image

blur_image=cv.GaussianBlur(image,(7,7),cv.BORDER_DEFAULT)
cv.imshow("blur_image",blur_image)

cv.waitKey(0)
```
